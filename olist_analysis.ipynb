# Section 1: Customer & Payment Insights
### Which regions are the most active, and how do people prefer to pay

```python
customer_orders= customers.merge(orders, on='customer_id',how='inner')
#merging customers and orders datasets on customer_id to get customer details for each order
customer_orders.head()
```

```python
#now we filter number of orders by region
region_orders= customer_orders.groupby('customer_state')['order_id'].count().sort_values(ascending=False)
region_orders.plot(kind='bar', figsize=(10, 5))
# we can see that the most active region is SP by a huge amount, followed by RJ and MG. We can also deduct that regions after SE have practically no orders
```

### now we check how people prefer to pay most

```python
customer_orders_payments=customer_orders.merge(order_payments,on='order_id', how='left')
customer_orders_payments.head()
customer_orders_payments['payment_type'].value_counts().plot(kind='bar')
#we can see that the most preferred payment method is credit card which is expected since products are dilevered online followed by boleto which is a payment method in brazil
```

# Section 2: Order Composition
### What do typical orders look like, and who are our biggest sellers?

```python
average_items_per_order = order_items.groupby('order_id')['product_id'].count().mean()
print(average_items_per_order)
# the avrage number of items per order is 1.14 where most orders have 1 item only
```

### now to find who the biggest sellers are

```python
seller_revenue=order_items.groupby('seller_id')['price'].sum().sort_values(ascending=False)
seller_revenue.head(20).plot(kind='bar',figsize=(10, 5))
plt.title('Top 20 Sellers by Quantity Sold')
plt.ylabel('revenue')
#we can see the top 20 sellers by quantity sold
```

# Section 3: Product Category Performance

### What product categories are driving sales volume?

```python
product_items= products.merge(order_items,how='left',on='product_id')
category_sales= product_items.groupby('product_category_name')['price'].sum().sort_values(ascending=False).reset_index(name='total')
category_sales.head(5).plot(kind='bar',x='product_category_name',y='total',figsize=(10,5))
#We conclude that beleza_saude is the the highest selling category
```

### now we find which category has the highest avrage item price

```python
category_avrage_item_price=product_items.groupby('product_category_name')['price'].mean().sort_values(ascending=False).reset_index(name='avg item price')
category_avrage_item_price.head(5).plot(kind='bar',x='product_category_name',y='avg item price')
#we conclude the pcs have the highest avrage item price which makes sense because pcs are expensive
```

# Section 4: Data Quality & Completeness

## Do we have any blind spots in our data?

### first we check if there are customers with no orders

```python
customer_orders_check=customers.merge(orders,how='outer',on='customer_id')
print(customer_orders_check[customer_orders_check['order_id'].isnull()])
#all customers have orders
```

### now we check if there are orders with no matching order items

```python
order_items_check= orders.merge(order_items,how='left',on='order_id')
print(order_items_check[order_items_check['order_item_id'].isnull()])
# we can see there are 775 orders with no matching items
```

### Are there payments linked to orders that dont exist?

```python
order_payments_check=orders.merge(order_payments,how='right',on='order_id')
order_payments_check[order_payments_check['order_status'].isnull()]
# i used a right join so that if there were orders in the payement table that werent in the orders table there would be mull valure in the orders table
#now we conclude that there are no payments linked to orders that dont exist 
```
